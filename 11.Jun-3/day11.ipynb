```python
import numpy as np
```

### Create two matrices

```python
A = np.array([[1, 2],[3, 4]])
B = np.array([[5, 6],[7, 8]])

print("Matrix A:")
print(A)

print("Matrix B:")
print(B)
```

### Matrix Addition

```python
addition = A + B

print("Matrix Addition:")
print(addition)
```

### Matrix Multiplication

###### Element-wise multiplication

```python
element_multiply = A * B

print("Element-wise Multiplication:")
print(element_multiply)
```

##### Real matrix multiplication

```python
matrix_multiply = np.dot(A, B)

print("Matrix Multiplication:")
print(matrix_multiply)
```

##### or

```python
matrix_multiply = A @ B
print(matrix_multiply)
```

### Mean

```python
mean_A = np.mean(A)

print("Mean of Matrix A:", mean_A)
```

### Median

```python
median_A = np.median(A)

print("Median of Matrix A:", median_A)
```

### Standard Deviation

```python
std_A = np.std(A)

print("Standard Deviation of Matrix A:", std_A)
```

## Full Code Together

```python
import numpy as np

A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])

print("Matrix A:")
print(A)

print("\nMatrix B:")
print(B)

addition = A + B
print("\nMatrix Addition:")
print(addition)

element_multiply = A * B
print("\nElement-wise Multiplication:")
print(element_multiply)

matrix_multiply = np.dot(A, B)
print("\nMatrix Multiplication:")
print(matrix_multiply)

mean_A = np.mean(A)
median_A = np.median(A)
std_A = np.std(A)

print("\nMean of Matrix A:", mean_A)
print("Median of Matrix A:", median_A)
print("Standard Deviation of Matrix A:", std_A)
```
