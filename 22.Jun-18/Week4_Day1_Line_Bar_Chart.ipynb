## Week 4 Day 1: Visualization
#### Topic : Line Chart and Bar Chart using Matplotlib + Pandas

### Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Create Simple Sales Dataset

```python
data = {
    "Month": ["January", "February", "March", "April", "May", "June"],
    "Sales": [12000, 15000, 18000, 14000, 22000, 25000],
    "Profit": [3000, 4000, 5000, 3500, 7000, 8000]
}

df = pd.DataFrame(data)

df
```

### Line Chart

```python
plt.figure(figsize=(8, 5))

plt.plot(df["Month"], df["Sales"], marker="o")

plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales Amount")
plt.grid(True)

plt.show()
```

### Bar Chart

```python
plt.figure(figsize=(8, 5))

plt.bar(df["Month"], df["Profit"])

plt.title("Monthly Profit Comparison")
plt.xlabel("Month")
plt.ylabel("Profit Amount")

plt.show()
```

### Sales and Profit Together

```python
plt.figure(figsize=(8, 5))

plt.plot(df["Month"], df["Sales"], marker="o", label="Sales")
plt.plot(df["Month"], df["Profit"], marker="o", label="Profit")

plt.title("Monthly Sales and Profit Trend")
plt.xlabel("Month")
plt.ylabel("Amount")
plt.legend()
plt.grid(True)

plt.show()
```
