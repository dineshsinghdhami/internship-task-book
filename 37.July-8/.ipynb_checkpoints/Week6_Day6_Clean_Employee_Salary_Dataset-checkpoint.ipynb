## Objective

The objective of this assignment is to clean a real-world employee salary dataset by handling missing values, duplicate records, outliers, incorrect data types, and inconsistent string formatting. The final cleaned dataset will be ready for analysis and machine learning tasks.

### Import Libraries

```python
import pandas as pd
import numpy as np
```

### Load Datasets

```python
df = pd.read_csv("employee_salary_dirty_dataset.csv")
df
```

### Dataset Information

```python
df.info()
```

```python
df.shape
```

### Check Missing Values

##### Fill Salary

###### First remove commas:

```python
df["Salary"] = df["Salary"].astype(str).str.replace(",", "")
```

###### Convert numeric:

```python
df["Salary"] = pd.to_numeric(df["Salary"], errors="coerce")
```

###### Fill missing salary:

```python
df["Salary"] = df["Salary"].fillna(df["Salary"].median())
```

### Fill Experience

```python
df["Experience_Years"] = pd.to_numeric(df["Experience_Years"], errors="coerce")
df["Experience_Years"] = df["Experience_Years"].fillna(df["Experience_Years"].mean())
```

### Check Duplicates

```python
df.duplicated().sum()
```

### Remove Duplicates

```python
df = df.drop_duplicates()
```

### Fix Data Types

```python
df = df.copy()
df["Joining_Date"] = pd.to_datetime(df["Joining_Date"], errors="coerce")
```

```python
df.info()
```

### Detect Outliers using IQR

```python
Q1 = df["Salary"].quantile(0.25)
Q3 = df["Salary"].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

print(lower, upper)
```

### Find Outliers

```python
outliers = df[(df["Salary"] < lower) | (df["Salary"] > upper)]
outliers
```

### Handle Outliers using Capping

```python
df["Salary"] = np.where(df["Salary"] > upper, upper, df["Salary"])
df["Salary"] = np.where(df["Salary"] < lower, lower, df["Salary"])
```

### String Cleaning

##### Employee Name:

```python
df["Employee_Name"] = df["Employee_Name"].str.strip().str.title()
```

##### Department:

```python
df["Department"] = df["Department"].str.strip().str.title()
```

##### City:

```python
df["City"] = df["City"].str.strip().str.title()
```

#### Gender cleaning:

```python
df["Gender"] = df["Gender"].str.strip().str.title()

df["Gender"] = df["Gender"].replace({
    "M": "Male",
    "F": "Female"
})
```

##### Employment Type:

```python
df["Employment_Type"] = df["Employment_Type"].str.strip().str.title()
```

### Final Check

```python
df.info()
```

###### or

```python
df
```

### Save Clean Dataset

```python
df.to_csv("employee_salary_cleaned_dataset.csv", index=False)
```

## Insights

- Missing values successfully handled.
- Duplicate rows removed.
- Outliers detected and capped using IQR.
- Incorrect data types fixed.
- String formatting standardized.
- Final dataset is clean and ready for analysis.
