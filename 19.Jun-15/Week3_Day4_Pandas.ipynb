## Aggregation in Pandas.

### Create Dataset

```python
import pandas as pd

data = {
    "Name": ["Ram", "Hari", "Sita", "Gita", "Shyam"],
    "Age": [20, 21, 19, 22, 20],
    "Marks": [80, 75, 90, 65, 55],
    "Department": ["CS", "IT", "CS", "BCA", "IT"]
}

df = pd.DataFrame(data)

print(df)
```

### Basic Aggregation

#### Total marks:

```python
df["Marks"].sum()
```

#### Mean (Average)

```python
df["Marks"].mean()
```

#### Maximum

```python
df["Marks"].max()
```

#### Minimum

```python
df["Marks"].min()
```

#### Count

```python
df["Marks"].count()
```

### Aggregation with Grouping

#### Mean by Department

```python
df.groupby("Department")["Marks"].mean()
```

#### Sum by Department

```python
df.groupby("Department")["Marks"].sum()
```

#### Max by Department

```python
df.groupby("Department")["Marks"].max()
```

#### Max by Department

```python
df.groupby("Department")["Marks"].min()
```

### Multiple Aggregation at once

```python
df.groupby("Department")["Marks"].agg(["sum", "mean", "max", "min"])
# Instead of writing one by one: Use .agg()
```

## Full Practice Code

```python
import pandas as pd

data = {
    "Name": ["Ram", "Hari", "Sita", "Gita", "Shyam"],
    "Age": [20, 21, 19, 22, 20],
    "Marks": [80, 75, 90, 65, 55],
    "Department": ["CS", "IT", "CS", "BCA", "IT"]
}

df = pd.DataFrame(data)

print("Dataset:")
print(df)

print("\nTotal Marks:")
print(df["Marks"].sum())

print("\nAverage Marks:")
print(df["Marks"].mean())

print("\nMaximum Marks:")
print(df["Marks"].max())

print("\nMinimum Marks:")
print(df["Marks"].min())

print("\nCount:")
print(df["Marks"].count())

print("\nDepartment-wise Average Marks:")
print(df.groupby("Department")["Marks"].mean())

print("\nMultiple Aggregation:")
print(df.groupby("Department")["Marks"].agg(["sum", "mean", "max", "min"]))
```
