### Create Dataset

```python
import pandas as pd

data = {
    "Name": ["Ram", "Hari", "Sita", "Gita", "Shyam"],
    "Age": [20, 21, 19, 22, 20],
    "Marks": [80, 75, 90, 65, 55],
    "Department": ["CS", "IT", "CS", "BCA", "IT"]
}

df = pd.DataFrame(data)

print(df)
```

### Select Columns

#### Name column

```python
df["Name"]
```

#### Marks column

```python
df["Marks"]
```

#### multiple columns

```python
df[["Name", "Marks"]]
```

### Select Rows

#### Select first row

```python
df.iloc[0]
# iloc means select using row number/index.
```

#### Select third row

```python
df.iloc[2]
```

#### Select multiple rows

```python
df.iloc[0:3]
```

### Select Specific Row + Column

#### Get value at row 1 and column 2.

```python
df.iloc[1, 2]
```

### Filtering Data

#### Students with marks > 70

```python
df[df["Marks"] > 70]
```

#### Students with marks < 60

```python
df[df["Marks"] < 60]
```

#### Students from CS Department

```python
df[df["Department"] == "CS"]
```

#### Students with age > 20

```python
df[df["Age"] > 20]
```

### Multiple Conditions

##### Students with:
##### Marks > 70
##### AND
##### Department = CS

```python
df[(df["Marks"] > 70) & (df["Department"] == "CS")]
```

#### Marks > 85 OR Age < 20

```python
df[(df["Marks"] > 85) | (df["Age"] < 20)]
```

## Full Practice Code

```python
import pandas as pd

data = {
    "Name": ["Ram", "Hari", "Sita", "Gita", "Shyam"],
    "Age": [20, 21, 19, 22, 20],
    "Marks": [80, 75, 90, 65, 55],
    "Department": ["CS", "IT", "CS", "BCA", "IT"]
}

df = pd.DataFrame(data)

print("Full Dataset:")
print(df)

print("\nSelect Name Column:")
print(df["Name"])

print("\nSelect Name and Marks:")
print(df[["Name", "Marks"]])

print("\nFirst Row:")
print(df.iloc[0])

print("\nRows 0 to 2:")
print(df.iloc[0:3])

print("\nStudents with Marks > 70:")
print(df[df["Marks"] > 70])

print("\nStudents from CS Department:")
print(df[df["Department"] == "CS"])

print("\nMultiple Conditions:")
print(df[(df["Marks"] > 70) & (df["Department"] == "CS")])
```
