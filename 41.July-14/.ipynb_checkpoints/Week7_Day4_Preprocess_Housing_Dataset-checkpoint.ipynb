## Objective

The objective of this assignment is to perform complete preprocessing on the housing dataset. This includes checking missing values, handling duplicates, encoding categorical columns, scaling numerical features, and creating new engineered features to prepare the dataset for machine learning.

### Import Libraries

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

### Load Dataset

```python
df = pd.read_csv("housing_dataset_assignment.csv")
df.head()
```

### Dataset Information

```python
df.info()
```

```python
df.shape
```

### Step 1: Check Missing Values

```python
df.isnull().sum()
```

### Step 2: Check Duplicates

```python
df.duplicated().sum()
```

### Step 3: Label Encoding

```python
encoder = LabelEncoder()

df["Location"] = encoder.fit_transform(df["Location"])
df["House_Type"] = encoder.fit_transform(df["House_Type"])
df["Parking"] = encoder.fit_transform(df["Parking"])
```

```python
df.head()
```

### Step 4: Feature Scaling

```python
numerical_cols = ["Bedrooms", "Bathrooms", "Area_sqft", "Age_Years", "Price"]
```

```python
scaler = StandardScaler()
df[numerical_cols] = scaler.fit_transform(df[numerical_cols])
```

```python
df.head()
```

### Step 5: Feature Engineering

##### Create Total Rooms

```python
df["Total_Rooms"] = df["Bedrooms"] + df["Bathrooms"]
```

##### Create Big House

```python
df["Big_House"] = np.where(df["Area_sqft"] > 0, 1, 0)
```

##### Create New House

```python
df["New_House"] = np.where(df["Age_Years"] < 0, 1, 0)
```

### Step 6: Final Dataset

```python
df.head()
```

### Final Dataset Info

```python
df.info()
```

### Step 7: Save Processed Dataset

```python
df.to_csv("housing_preprocessed_dataset.csv", index=False)
```

## Conclusion

In this assignment, I performed complete preprocessing on the housing dataset. I checked missing values and duplicate records, encoded categorical columns, scaled numerical features using StandardScaler, and created new engineered features. The final dataset is now clean and ready for machine learning model development.
