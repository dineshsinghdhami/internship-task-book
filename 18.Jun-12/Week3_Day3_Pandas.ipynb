## Sorting + Grouping in Pandas.

```python
import pandas as pd
```

### Create Dataset

```python
import pandas as pd

data = {
    "Name": ["Ram", "Hari", "Sita", "Gita", "Shyam"],
    "Age": [20, 21, 19, 22, 20],
    "Marks": [80, 75, 90, 65, 55],
    "Department": ["CS", "IT", "CS", "BCA", "IT"]
}

df = pd.DataFrame(data)

print(df)
```

## Sorting

#### Sort by Marks (Ascending)

```python
df.sort_values(by="Marks")
```

#### Sort by Marks (Descending)

```python
df.sort_values(by="Marks", ascending=False)
```

#### Sort by Age

```python
df.sort_values(by="Age")
```

### Sort by Multiple Columns

##### Sort by:
##### Department
##### Marks

```python
df.sort_values(by=["Department", "Marks"])
```

## Grouping

#### Group by Department

```python
df.groupby("Department")
```

#### Count students in each department

```python
df.groupby("Department").size()
```

#### Average marks by department

```python
df.groupby("Department")["Marks"].mean()
```

#### Maximum marks by department

```python
df.groupby("Department")["Marks"].max()
```

#### Minimum marks by department

```python
df.groupby("Department")["Marks"].min()
```

#### Sum of marks by department

```python
df.groupby("Department")["Marks"].sum()
```

## Full Practice Code

```python
import pandas as pd

data = {
    "Name": ["Ram", "Hari", "Sita", "Gita", "Shyam"],
    "Age": [20, 21, 19, 22, 20],
    "Marks": [80, 75, 90, 65, 55],
    "Department": ["CS", "IT", "CS", "BCA", "IT"]
}

df = pd.DataFrame(data)

print("Original Dataset:")
print(df)

print("\nSorted by Marks Ascending:")
print(df.sort_values(by="Marks"))

print("\nSorted by Marks Descending:")
print(df.sort_values(by="Marks", ascending=False))

print("\nSorted by Age:")
print(df.sort_values(by="Age"))

print("\nStudents Count by Department:")
print(df.groupby("Department").size())

print("\nAverage Marks by Department:")
print(df.groupby("Department")["Marks"].mean())

print("\nMaximum Marks by Department:")
print(df.groupby("Department")["Marks"].max())
```
