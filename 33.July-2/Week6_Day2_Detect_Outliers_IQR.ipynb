## Objective

The objective of this task is to detect outliers in datasets using the Interquartile Range (IQR) method. Outliers are abnormal or extreme values that can negatively affect data quality, analysis, and machine learning performance.

### Import Libraries

```python
import pandas as pd
import numpy as np
```

### Load Dataset

```python
df = pd.read_csv("outlier_detection_dataset.csv")
df.head()
```

### Understand Dataset

###### Check dataset shape:

```python
df.shape
```

###### Check info:

```python
df.info()
```

###### Check salary column:

```python
df["Salary"]
```

### Calculate Quartiles

```python
Q1 = df["Salary"].quantile(0.25)
Q3 = df["Salary"].quantile(0.75)

print("Q1:", Q1)
print("Q3:", Q3)
```

### Calculate IQR

```python
IQR = Q3 - Q1
print("IQR:", IQR)
```

### Calculate Bounds

```python
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)
```

### Detect Outliers

```python
outliers = df[(df["Salary"] < lower_bound) | (df["Salary"] > upper_bound)]
outliers
```

### Count Outliers

```python
outliers.shape
```

###### OR

```python
len(outliers)
```
