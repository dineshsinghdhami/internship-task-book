## Objective
The objective of this task is to learn how to handle missing values using forward fill and backward fill methods in Pandas. These techniques are useful when working with sequential or time-series data.

### Import Libraries

```python
import pandas as pd
import numpy as np
```

### Create Sample Dataset

```python
data = {
    "Day": [1, 2, 3, 4, 5, 6],
    "Sales": [1000, None, None, 1500, None, 2000]
}

df = pd.DataFrame(data)
df
```

#### Task 1: Forward Fill

```python
df.ffill()
###### Row 2 takes value from Row 1
###### Row 3 also takes previous filled value
###### Row 5 takes value from Row 4
```

#### Task 2: Backward Fill

```python
df.bfill()
```

#### Task 3: Apply on Real Dataset

```python
df = pd.read_csv("customer_data.csv")
df
```

###### Check missing values:

```python
df.isnull().sum()
```

###### Apply Forward Fill

```python
df.ffill()
```

###### Apply Backward Fill

```python
df.bfill()
```

### When to Use?
##### Use Forward Fill:
Time-series data, Stock prices, Daily sales records
##### Use Backward Fill:
When next value is more relevant

## Conclusion
In this task, I learned how to handle missing values using forward fill and backward fill methods in Pandas. Forward fill replaces missing values using previous valid values, while backward fill uses next valid values. These techniques are useful for sequential and time-series datasets.
