## Objective
The objective of this task is to understand Scatter Plot visualization using Python. Scatter plot helps analyze the relationship between two numerical variables and identify trends, patterns, and correlations.

### Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

##### Create Dataset

```python
study_hours = [1, 2, 3, 4, 5, 6, 7, 8]
marks = [35, 40, 50, 60, 65, 75, 85, 95]
```

#### Scatter Plot Code

```python
plt.figure(figsize=(8,5))

plt.scatter(study_hours, marks)

plt.title("Study Hours vs Student Marks")
plt.xlabel("Study Hours")
plt.ylabel("Marks")

plt.grid(True)
plt.show()
```
