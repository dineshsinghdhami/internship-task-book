### Import Libraries

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

### Load Dataset

```python
df = pd.read_csv("housing_dataset_week7.csv")
df.head()
```

### Select Numerical Columns

```python
numerical_cols = ["Bedrooms", "Bathrooms", "Area_sqft", "Age_Years", "Price"]

df[numerical_cols].head()
```

### Part 1: Standardization

##### Apply Standardization

```python
scaler = StandardScaler()

df_standardized = df.copy()
df_standardized[numerical_cols] = scaler.fit_transform(df[numerical_cols])

df_standardized.head()
```

##### Check Standardized Values

```python
df_standardized[numerical_cols].describe()
```

### Part 2: Normalization

##### Apply Normalization

```python
minmax = MinMaxScaler()

df_normalized = df.copy()
df_normalized[numerical_cols] = minmax.fit_transform(df[numerical_cols])

df_normalized.head()
```

##### Check Normalized Values

```python
df_normalized[numerical_cols].describe()
```

### Compare Original vs Scaled

```python
print("Original:")
print(df[numerical_cols].head())

print("\nStandardized:")
print(df_standardized[numerical_cols].head())

print("\nNormalized:")
print(df_normalized[numerical_cols].head())
```

#### Save Datasets

```python
df_standardized.to_csv("housing_standardized.csv", index=False)
df_normalized.to_csv("housing_normalized.csv", index=False)
```

## Conclusion

In this task, I learned feature scaling techniques using Standardization and Normalization. Standardization transformed numerical features to have mean close to 0 and standard deviation close to 1, while Normalization scaled values between 0 and 1. These preprocessing techniques improve machine learning model performance and stability.
