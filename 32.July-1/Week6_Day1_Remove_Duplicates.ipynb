## Objective

The objective of this task is to identify and remove duplicate records from datasets using Pandas. Duplicate records can negatively affect data quality, analytics, and machine learning performance.

### Import Libraries

```python
import pandas as pd
import numpy as np
```

#### Load Dataset

```python
df = pd.read_csv("duplicates_practice_dataset.csv")
df.head()
```

#### Dataset Shape

```python
df.shape
```

### Detect Duplicate Records

```python
df.duplicated()
```

### Count Duplicates

```python
df.duplicated().sum()
```

### View Duplicate Rows

```python
duplicates = df[df.duplicated()]
duplicates
```

### Remove Duplicates

```python
df_cleaned = df.drop_duplicates()
df_cleaned
```

### Compare Shape

```python
print("Before:", df.shape)
print("After:", df_cleaned.shape)
```

### Verify Final Dataset

```python
df_cleaned.duplicated().sum()
```

## Conclusion

In this task, I learned how to detect and remove duplicate records using Pandas. I used duplicated() to identify duplicate rows and drop_duplicates() to remove them. Removing duplicates improves data quality and ensures accurate analysis.
