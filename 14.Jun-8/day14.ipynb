## Random Dataset Generation + Array Filtering using NumPy

```python
import numpy as np
```

### Generate random dataset

```python
marks = np.random.randint(30, 101, 20)

print("Student Marks:")
print(marks)
```

### Filter marks greater than 80

```python
high_marks = marks[marks > 80]

print("Marks greater than 80:")
print(high_marks)
```

### Filter marks less than 40

```python
failed_marks = marks[marks < 40]

print("Failed Marks:")
print(failed_marks)
```

### Filter marks between 50 and 70

```python
average_marks = marks[(marks >= 50) & (marks <= 70)]

print("Marks between 50 and 70:")
print(average_marks)
```

### Count passed students

```python
passed_students = np.sum(marks >= 40)

print("Number of passed students:", passed_students)
```

### Count failed students

```python
failed_students = np.sum(marks < 40)

print("Number of failed students:", failed_students)
```

### Find highest, lowest, average marks

```python
print("Highest Marks:", np.max(marks))
print("Lowest Marks:", np.min(marks))
print("Average Marks:", np.mean(marks))
```

## Full code

```python
import numpy as np

marks = np.random.randint(30, 101, 20)

print("Student Marks:")
print(marks)

high_marks = marks[marks > 80]
print("\nMarks greater than 80:")
print(high_marks)

failed_marks = marks[marks < 40]
print("\nFailed Marks:")
print(failed_marks)

average_marks = marks[(marks >= 50) & (marks <= 70)]
print("\nMarks between 50 and 70:")
print(average_marks)

passed_students = np.sum(marks >= 40)
failed_students = np.sum(marks < 40)

print("\nNumber of passed students:", passed_students)
print("Number of failed students:", failed_students)

print("\nHighest Marks:", np.max(marks))
print("Lowest Marks:", np.min(marks))
print("Average Marks:", np.mean(marks))
```
