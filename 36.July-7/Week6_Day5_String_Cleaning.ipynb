## Objective

The objective of this task is to clean text-based columns in a dataset using Pandas. String cleaning helps remove extra spaces, fix inconsistent capitalization, and standardize categorical values for better analysis.

### Import Library

```python
import pandas as pd
```

### Load Dataset

```python
df = pd.read_csv("string_cleaning_dataset.csv")
df
```

### Check Dataset Info

```python
df.info()
```

### Check Unique Values Before Cleaning

```python
df["Gender"].unique()
```

```python
df["City"].unique()
```

```python
df["Membership_Type"].unique()
```

### Remove Extra Spaces

```python
df["Customer_Name"] = df["Customer_Name"].str.strip()
df["Gender"] = df["Gender"].str.strip()
df["City"] = df["City"].str.strip()
df["Email"] = df["Email"].str.strip()
df["Membership_Type"] = df["Membership_Type"].str.strip()
df["Payment_Method"] = df["Payment_Method"].str.strip()
```

### Fix Capitalization

```python
df["Customer_Name"] = df["Customer_Name"].str.title()
df["Gender"] = df["Gender"].str.title()
df["City"] = df["City"].str.title()
df["Membership_Type"] = df["Membership_Type"].str.title()
df["Payment_Method"] = df["Payment_Method"].str.title()
```

### Convert Email to Lowercase

```python
df["Email"] = df["Email"].str.lower()
```

### Replace Short Forms

```python
df["Gender"] = df["Gender"].replace({
    "M": "Male",
    "F": "Female"
})
```

### Fix Payment Method Format

```python
df["Payment_Method"] = df["Payment_Method"].replace({
    "Esewa": "eSewa",
    "Bank Transfer": "Bank Transfer"
})
```

### Check Unique Values After Cleaning

```python
df["Gender"].unique()
```

```python
df["City"].unique()
```

```python
df["Membership_Type"].unique()
```

```python
df["Payment_Method"].unique()
```

### Final Cleaned Dataset

```python
df
```

### Save Cleaned Dataset

```python
df.to_csv("cleaned_string_dataset.csv", index=False)
```
