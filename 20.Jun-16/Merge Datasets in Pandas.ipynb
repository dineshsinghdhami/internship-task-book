## Merge Datasets in Pandas.
##### This is a very important real-world topic because in companies data usually comes from multiple files/tables.

### Import Pandas

```python
import pandas as pd
```

### Create Dataset 1

```python
student_info = {
    "Student_ID": [1, 2, 3],
    "Name": ["Ram", "Hari", "Sita"],
    "Age": [20, 21, 19]
}

df1 = pd.DataFrame(student_info)

print("Student Info Dataset:")
print(df1)
```

### Create Dataset 2

```python
marks_data = {
    "Student_ID": [1, 2, 3],
    "Marks": [80, 75, 90]
}

df2 = pd.DataFrame(marks_data)

print("Marks Dataset:")
print(df2)
```

### Merge Datasets

```python
merged_df = pd.merge(df1, df2, on="Student_ID")
print(merged_df)
```

### Types of Merge
##### There are 4 important types:
##### 1. Inner Join
##### 2. Left Join
##### 3. Right Join
##### 4. Outer Join

### #Inner Merge (Common Rows Only)

```python
pd.merge(df1, df2, on="Student_ID", how="inner")
```

### #Left Merge

```python
pd.merge(df1, df2, on="Student_ID", how="left")
```

### #Right Merge

```python
pd.merge(df1, df2, on="Student_ID", how="right")
```

### #Outer Merge

```python
pd.merge(df1, df2, on="Student_ID", how="outer")
```

## Full Practice Code

```python
import pandas as pd

student_info = {
    "Student_ID": [1, 2, 3],
    "Name": ["Ram", "Hari", "Sita"],
    "Age": [20, 21, 19]
}

marks_data = {
    "Student_ID": [1, 2, 3],
    "Marks": [80, 75, 90]
}

df1 = pd.DataFrame(student_info)
df2 = pd.DataFrame(marks_data)

print("Student Info:")
print(df1)

print("\nMarks Data:")
print(df2)

merged_df = pd.merge(df1, df2, on="Student_ID")

print("\nMerged Dataset:")
print(merged_df)
```
