## Objective
The objective of this task is to understand Histogram and Pie Chart visualization techniques using Python. Histogram helps analyze data distribution, while Pie Chart helps visualize proportions and percentage contribution.

### Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Create Dataset

```python
student_marks = [45, 56, 67, 78, 89, 90, 76, 65, 54, 43, 81, 72, 60, 68, 75]

sales_data = {
    "Category": ["Electronics", "Clothing", "Food", "Books"],
    "Sales": [40000, 25000, 20000, 15000]
}
```

### Histogram

```python
plt.figure(figsize=(8,5))

plt.hist(student_marks, bins=5, edgecolor='black')

plt.title("Distribution of Student Marks")
plt.xlabel("Marks")
plt.ylabel("Number of Students")

plt.show()
```

### Pie Chart

```python
df = pd.DataFrame(sales_data)

plt.figure(figsize=(6,6))

plt.pie(df["Sales"], labels=df["Category"], autopct="%1.1f%%")

plt.title("Sales Distribution by Category")

plt.show()
```
