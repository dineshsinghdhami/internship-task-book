## Final Full Code

```python
import pandas as pd

data = {
    "Name": ["Ram", "Hari", "Sita"],
    "Age": [20, 21, 19],
    "Marks": [80, 75, 90]
}

df = pd.DataFrame(data)

print("DataFrame:")
print(df)

print("\nFirst 5 rows:")
print(df.head())

print("\nLast 5 rows:")
print(df.tail())

print("\nInfo:")
print(df.info())

print("\nStatistics:")
print(df.describe())
```
