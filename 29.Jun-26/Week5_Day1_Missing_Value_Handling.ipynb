## Objective
The objective of this task is to learn how to identify and handle missing values in datasets using Pandas. Missing value handling is one of the most important steps in data cleaning and preprocessing.

### Import Libraries

```python
import pandas as pd
import numpy as np
```

### Create Dataset with Missing Values

```python
data = {
    "Name": ["Ram", "Hari", "Sita", "Gita", "Rita"],
    "Age": [22, None, 24, 23, None],
    "Salary": [50000, 45000, None, 55000, 52000]
}

df = pd.DataFrame(data)
df
```

#### Task 1: Find Null Values

###### Check Null Values

```python
df.isnull()
```

###### Count Null Values

```python
df.isnull().sum()
```

###### Dataset Info

```python
df.info()
```

### Task 2: Remove Null Rows and Columns

###### Remove Rows with Null Values

```python
df.dropna()
```

###### Remove Columns with Null Values

```python
df.dropna(axis=1)
```

### Task 3: Fill Missing Values

###### Fill Using Mean

```python
df["Age"] = df["Age"].fillna(df["Age"].mean())
df
```

###### Fill Using Median

```python
df["Salary"] = df["Salary"].fillna(df["Salary"].median())
df
```

###### Fill Using Mode

```python
df["Age"] = df["Age"].fillna(df["Age"].mode()[0])
df
```
