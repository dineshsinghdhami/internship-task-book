## Project: Student Performance Dataset Analysis

### Create Dataset

```python
import pandas as pd

data = {
    "Name": ["Ram", "Hari", "Sita", "Gita", "Shyam", "Rita"],
    "Age": [20, 21, 19, 22, 20, 21],
    "Department": ["CS", "IT", "CS", "BCA", "IT", "CS"],
    "Attendance": [90, 85, 95, 70, 60, 88],
    "Marks": [80, 75, 92, 65, 50, 85]
}

df = pd.DataFrame(data)

print(df)
```

### Basic Information

```python
print(df.head())
```

```python
print(df.info())
```

```python
print(df.describe())
```

### Find Top Students

#### Students with marks > 80

```python
top_students = df[df["Marks"] > 80]
print(top_students)
```

#### Marks below 60

```python
weak_students = df[df["Marks"] < 60]
print(weak_students)
```

### Sort by Marks

##### Highest marks first

```python
sorted_data = df.sort_values(by="Marks", ascending=False)
print(sorted_data)
```

##### Average Marks

```python
average_marks = df["Marks"].mean()
print("Average Marks:", average_marks)
```

##### Highest and Lowest Marks

```python
print("Highest Marks:", df["Marks"].max())
print("Lowest Marks:", df["Marks"].min())
```

##### Department-wise Average Marks

```python
dept_avg = df.groupby("Department")["Marks"].mean()
print(dept_avg)
```

##### Attendance vs Marks

###### Students with attendance > 85

```python
good_attendance = df[df["Attendance"] > 85]
print(good_attendance)
```

### Full Code

```python
import pandas as pd

data = {
    "Name": ["Ram", "Hari", "Sita", "Gita", "Shyam", "Rita"],
    "Age": [20, 21, 19, 22, 20, 21],
    "Department": ["CS", "IT", "CS", "BCA", "IT", "CS"],
    "Attendance": [90, 85, 95, 70, 60, 88],
    "Marks": [80, 75, 92, 65, 50, 85]
}

df = pd.DataFrame(data)

print("Student Dataset:")
print(df)

print("\nDataset Info:")
print(df.info())

print("\nStatistics:")
print(df.describe())

print("\nTop Students:")
print(df[df["Marks"] > 80])

print("\nWeak Students:")
print(df[df["Marks"] < 60])

print("\nSorted by Marks:")
print(df.sort_values(by="Marks", ascending=False))

print("\nAverage Marks:")
print(df["Marks"].mean())

print("\nHighest Marks:")
print(df["Marks"].max())

print("\nLowest Marks:")
print(df["Marks"].min())

print("\nDepartment-wise Average Marks:")
print(df.groupby("Department")["Marks"].mean())
```
