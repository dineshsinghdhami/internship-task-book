## Student Performance Analysis

### Import Libraries

```python
import pandas as pd
```

### Read CSV

```python
df = pd.read_csv("StudentsPerformance.csv")
```

### Check Dataset

```python
print(df.head())
```

#### Basic Info

```python
print(df.shape)
```

```python
print(df.columns)
```

```python
print(df.info())
```

#### Statistics

```python
print(df.describe())
```

#### Check Missing Values

```python
print(df.isnull().sum())
```

#### Top Students

###### Students with G3 > 15

```python
top_students = df[df["G3"] > 15]
print(top_students[["sex", "age", "studytime", "G1", "G2", "G3"]])
```

#### Weak Students

###### Students with G3 < 8

```python
weak_students = df[df["G3"] < 8]
print(weak_students[["sex", "age", "studytime", "G1", "G2", "G3"]])
```

#### Average Final Grade

```python
print("Average Final Grade:", df["G3"].mean())
```

#### Highest and Lowest Grade

```python
print("Highest Final Grade:", df["G3"].max())
print("Lowest Final Grade:", df["G3"].min())
```

#### Gender-wise Performance

```python
print(df.groupby("sex")["G3"].mean())
```

#### Study Time vs Final Grade

```python
print(df.groupby("studytime")["G3"].mean())
```

#### Failures vs Grade

```python
print(df.groupby("failures")["G3"].mean())
```

#### Absences vs Grade

###### Students with high absences:

```python
high_absence = df[df["absences"] > 10]
print(high_absence[["absences", "G3"]])
```

###### Check average:

```python
print(df.groupby("absences")["G3"].mean())
```

#### Sort by Final Grade

```python
top = df.sort_values(by="G3", ascending=False)
print(top[["sex", "age", "studytime", "G3"]].head(10))
```

#### Create Performance Category

```python
def performance(score):
    if score >= 16:
        return "Excellent"
    elif score >= 12:
        return "Good"
    elif score >= 8:
        return "Average"
    else:
        return "Poor"

df["Performance"] = df["G3"].apply(performance)
```

###### Check:

```python
print(df[["G3", "Performance"]].head())
```

###### Count categories:

```python
print(df["Performance"].value_counts())
```
