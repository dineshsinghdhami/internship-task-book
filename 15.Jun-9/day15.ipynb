## Project: Sales Analyzer using NumPy

```python
import numpy as np
```

### Create sales data

```python
sales = np.array([
    [1200, 1500, 1000, 1800, 2000],
    [1300, 1600, 1100, 1700, 2100],
    [1250, 1550, 1050, 1900, 2200],
    [1400, 1650, 1150, 2000, 2300],
    [1500, 1700, 1200, 2100, 2400],
    [1600, 1800, 1300, 2200, 2500],
    [1700, 1900, 1400, 2300, 2600]
])

print("Sales Data:")
print(sales)
```

### Total sales of all products

```python
total_sales = np.sum(sales)

print("Total Sales:", total_sales)
```

### Daily sales

```python
daily_sales = np.sum(sales, axis=1)

print("Daily Sales:")
print(daily_sales)
```

### Product-wise sales

```python
product_sales = np.sum(sales, axis=0)

print("Product-wise Sales:")
print(product_sales)
```

### Average sales

```python
average_sales = np.mean(sales)

print("Average Sales:", average_sales)
```

### Highest and lowest sales

```python
highest_sale = np.max(sales)
lowest_sale = np.min(sales)

print("Highest Sale:", highest_sale)
print("Lowest Sale:", lowest_sale)
```

### Best selling product

```python
best_product_index = np.argmax(product_sales)

print("Best Selling Product: Product", best_product_index + 1)
print("Sales Amount:", product_sales[best_product_index])
```

### Worst selling product

```python
worst_product_index = np.argmin(product_sales)

print("Worst Selling Product: Product", worst_product_index + 1)
print("Sales Amount:", product_sales[worst_product_index])
```

### Filter sales greater than 2000

```python
high_sales = sales[sales > 2000]

print("Sales greater than 2000:")
print(high_sales)
```

## Full Code

```python
import numpy as np

sales = np.array([
    [1200, 1500, 1000, 1800, 2000],
    [1300, 1600, 1100, 1700, 2100],
    [1250, 1550, 1050, 1900, 2200],
    [1400, 1650, 1150, 2000, 2300],
    [1500, 1700, 1200, 2100, 2400],
    [1600, 1800, 1300, 2200, 2500],
    [1700, 1900, 1400, 2300, 2600]
])

print("Sales Data:")
print(sales)

total_sales = np.sum(sales)
print("\nTotal Sales:", total_sales)

daily_sales = np.sum(sales, axis=1)
print("\nDaily Sales:")
print(daily_sales)

product_sales = np.sum(sales, axis=0)
print("\nProduct-wise Sales:")
print(product_sales)

average_sales = np.mean(sales)
print("\nAverage Sales:", average_sales)

highest_sale = np.max(sales)
lowest_sale = np.min(sales)

print("\nHighest Sale:", highest_sale)
print("Lowest Sale:", lowest_sale)

best_product_index = np.argmax(product_sales)
worst_product_index = np.argmin(product_sales)

print("\nBest Selling Product: Product", best_product_index + 1)
print("Best Product Sales:", product_sales[best_product_index])

print("\nWorst Selling Product: Product", worst_product_index + 1)
print("Worst Product Sales:", product_sales[worst_product_index])

high_sales = sales[sales > 2000]
print("\nSales greater than 2000:")
print(high_sales)
```
