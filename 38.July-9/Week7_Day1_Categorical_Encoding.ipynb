### Import Libraries

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

### Load Dataset

```python
df = pd.read_csv("housing_dataset_week7.csv")
df.head()
```

### Check Dataset Information

```python
df.info()
```

### Check Categorical Columns

```python
df.select_dtypes(include="object").head()
```

## Part 1: Label Encoding

### View Unique Values

```python
print(df["Location"].unique())
print(df["House_Type"].unique())
print(df["Parking"].unique())
```

### Apply Label Encoding

```python
label_encoder = LabelEncoder()

df["Location_Label"] = label_encoder.fit_transform(df["Location"])
df["House_Type_Label"] = label_encoder.fit_transform(df["House_Type"])
df["Parking_Label"] = label_encoder.fit_transform(df["Parking"])

df.head()
```

### Check Encoded Columns

```python
df[["Location", "Location_Label", "House_Type", "House_Type_Label", "Parking", "Parking_Label"]].head()
```

## Part 2: One Hot Encoding

### Apply One Hot Encoding

```python
df_onehot = pd.get_dummies(df, columns=["Location", "House_Type", "Parking"], drop_first=True)

df_onehot.head()
```

### Check Final Columns

```python
df_onehot.columns
```

### Compare Dataset Shape

```python
print("Original Dataset Shape:", df.shape)
print("One Hot Encoded Dataset Shape:", df_onehot.shape)
```

### Save Encoded Dataset

```python
df_onehot.to_csv("encoded_housing_dataset.csv", index=False)
```

## Conclusion

In this task, I learned how to convert categorical data into numerical format using Label Encoding and One Hot Encoding. Label Encoding assigns numeric labels to categories, while One Hot Encoding creates separate binary columns for each category. These encoding techniques are important preprocessing steps before applying machine learning models.
