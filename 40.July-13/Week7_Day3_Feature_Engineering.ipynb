### Import Libraries

```python
import pandas as pd
import numpy as np
```

### Load Dataset

```python
df = pd.read_csv("housing_dataset_week7.csv")
df.head()
```

### Dataset Info

```python
df.info()
```

### Step 1: Create Price per Sqft

```python
df["Price_per_sqft"] = df["Price"] / df["Area_sqft"]
df.head()
```

### Step 2: Create Total Rooms

```python
df["Total_Rooms"] = df["Bedrooms"] + df["Bathrooms"]
df.head()
```

### Step 3: Create House Age Category
###### 0–5 → New
###### 6–15 → Medium
###### 16+ → Old

```python
def age_category(age):
    if age <= 5:
        return "New"
    elif age <= 15:
        return "Medium"
    else:
        return "Old"

df["House_Age_Category"] = df["Age_Years"].apply(age_category)
df.head()
```

### Step 4: Create Luxury House Feature

###### If Price > 30,000,000 → Luxury

```python
df["Luxury_House"] = np.where(df["Price"] > 30000000, 1, 0)
df.head()
```

### Step 5: Create Big House Feature

```python
df["Big_House"] = np.where(df["Area_sqft"] > 2000, 1, 0)
df.head()
```

### Check New Columns

```python
df.columns
```

### Final Dataset

```python
df.head()
```

### Save Dataset

```python
df.to_csv("housing_feature_engineered.csv", index=False)
```
