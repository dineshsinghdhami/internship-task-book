## Objective

The objective of this task is to handle outliers in datasets using different techniques such as removal and capping. Handling outliers improves data quality and ensures better analysis and machine learning model performance.

### Import Libraries

```python
import pandas as pd
import numpy as np
```

### Load Dataset

```python
df = pd.read_csv("handle_outliers_dataset.csv")
df.head()
```

### Check Dataset Information

```python
df.info()
```

```python
df.shape
```

### View Salary Statistics

```python
df["Salary"].describe()
```

### Calculate Q1, Q3, and IQR

```python
Q1 = df["Salary"].quantile(0.25)
Q3 = df["Salary"].quantile(0.75)

IQR = Q3 - Q1

print("Q1:", Q1)
print("Q3:", Q3)
print("IQR:", IQR)
```

### Calculate Lower and Upper Bound

```python
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)
```

### Detect Outliers

```python
outliers = df[(df["Salary"] < lower_bound) | (df["Salary"] > upper_bound)]
outliers
```

### Count Outliers

```python
print("Total Outliers:", len(outliers))
```

### Method 1 - Remove Outliers

```python
df_removed = df[(df["Salary"] >= lower_bound) & (df["Salary"] <= upper_bound)]
df_removed
```

### Compare Shape Before and After Removing

```python
print("Original Dataset Shape:", df.shape)
print("After Removing Outliers:", df_removed.shape)
```

### Method 2 - Cap Outliers

```python
df_capped = df.copy()

df_capped["Salary"] = np.where(
    df_capped["Salary"] > upper_bound,
    upper_bound,
    df_capped["Salary"]
)

df_capped["Salary"] = np.where(
    df_capped["Salary"] < lower_bound,
    lower_bound,
    df_capped["Salary"]
)

df_capped
```

### Compare Original and Capped Salary

```python
comparison = pd.DataFrame({
    "Original_Salary": df["Salary"],
    "Capped_Salary": df_capped["Salary"]
})

comparison
```

### Check Salary Statistics After Capping

```python
df_capped["Salary"].describe()
```

### Save Cleaned Dataset

```python
df_capped.to_csv("handled_outliers_dataset.csv", index=False)
```

## Conclusion

In this task, I learned how to handle outliers using IQR-based techniques. First, outliers were detected from the Salary column using lower and upper bounds. Then, outliers were handled using two methods: removing outlier records and capping extreme values. Capping was useful because it preserved all records while reducing the effect of abnormal values.
