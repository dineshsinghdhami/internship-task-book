## Objective
The objective of this task is to understand correlation heatmap visualization using Python. Heatmaps help identify relationships between multiple numerical variables and show how strongly features are correlated with each other.

#### Install Seaborn

```python
pip install seaborn
```

### Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Create Dataset
##### Student dataset

```python
data = {
    "Study_Hours": [2, 3, 4, 5, 6, 7, 8],
    "Attendance": [60, 65, 70, 75, 80, 85, 90],
    "Assignment": [50, 55, 60, 70, 75, 80, 90],
    "Marks": [45, 50, 60, 70, 75, 85, 95]
}

df = pd.DataFrame(data)
df
```

### Correlation Matrix

```python
correlation = df.corr()
correlation
```

### Heatmap Code

```python
plt.figure(figsize=(8,6))

sns.heatmap(correlation, annot=True, cmap="coolwarm")

plt.title("Correlation Heatmap")
plt.show()
```
