## Objective
The objective of this assignment is to clean a customer dataset by identifying missing values, handling null records, filling missing values using suitable techniques, and preparing the dataset for further analysis.

### Import Libraries

```python
import pandas as pd
import numpy as np
```

### Load Dataset

```python
df = pd.read_csv("customer_data.csv")
df
```

### Check Dataset Information

```python
df.info()
```

```python
df.shape
```

```python
df.head()
```

### Find Missing Values

```python
df.isnull()
```

```python
df.isnull().sum()
```

### Check Duplicate Records

```python
df.duplicated().sum()
```

### Remove Duplicate Records

```python
df = df.drop_duplicates()
df
```

### Fill Missing Age with Mean

```python
df.loc[:, "Age"] = df["Age"].fillna(df["Age"].mean())
df
```

### Fill Missing Salary with Median

```python
df.loc[:, "Salary"] = df["Salary"].fillna(df["Salary"].median())
df
```

### Fill Missing City with Mode

```python
df.loc[:, "City"] = df["City"].fillna(df["City"].mode()[0])
df
```

### Fill Missing Purchase Amount with Median

```python
df.loc[:, "Purchase_Amount"] = df["Purchase_Amount"].fillna(df["Purchase_Amount"].median())
df
```

### Check Missing Values Again

```python
df.isnull().sum()
```

### Save Cleaned Dataset

```python
df.to_csv("cleaned_customer_data.csv", index=False)
```

### Final Cleaned Dataset

```python
df
```

## Insights
- Missing values were found in Age, City, Salary, and Purchase Amount columns.
- Age was filled using mean value.
- Salary and Purchase Amount were filled using median value.
- City was filled using mode value.
- Duplicate records were checked and removed.
- Final cleaned dataset is ready for analysis.
