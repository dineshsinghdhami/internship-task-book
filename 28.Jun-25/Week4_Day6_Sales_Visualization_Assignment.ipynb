## Objective
The objective of this assignment is to perform visualization on a sales dataset using Python. Different chart types such as line chart, bar chart, histogram, and pie chart will be used to analyze sales trends, profit distribution, and category performance.

### Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Create Sales Dataset

```python
data = {
    "Month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun"],
    "Sales": [12000, 15000, 18000, 14000, 22000, 25000],
    "Profit": [3000, 4000, 5000, 3500, 7000, 8000],
    "Orders": [100, 120, 150, 110, 180, 200]
}

df = pd.DataFrame(data)
df
```

### Line Chart (Sales Trend)

```python
plt.figure(figsize=(8,5))
plt.plot(df["Month"], df["Sales"], marker='o')
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

### Bar Chart (Profit Comparison)

```python
plt.figure(figsize=(8,5))
plt.bar(df["Month"], df["Profit"])
plt.title("Monthly Profit Comparison")
plt.xlabel("Month")
plt.ylabel("Profit")
plt.show()
```

### Histogram (Orders Distribution)

```python
plt.figure(figsize=(8,5))
plt.hist(df["Orders"], bins=5, edgecolor="black")
plt.title("Orders Distribution")
plt.xlabel("Orders")
plt.ylabel("Frequency")
plt.show()
```

### Pie Chart (Sales Contribution)

```python
plt.figure(figsize=(6,6))
plt.pie(df["Sales"], labels=df["Month"], autopct="%1.1f%%")
plt.title("Sales Contribution by Month")
plt.show()
```

## Insights
- Sales increased steadily from January to June.
- Highest sales were recorded in June.
- Profit also increased with sales.
- Order volume improved significantly in later months.
