## Objective

The objective of this task is to identify incorrect data types in datasets and convert them into appropriate formats such as numeric, datetime, and categorical types for better analysis and preprocessing.

### Import Libraries

```python
import pandas as pd
import numpy as np
```

### Load Datasets

```python
df = pd.read_csv("wrong_datatypes_dataset.csv")
df
```

### Check Dataset Info

```python
df.info()
```

### Fix Age Column

```python
df["Age"] = pd.to_numeric(df["Age"], errors="coerce")
```

```python
df
```

### Fix Salary Column

###### First remove commas:

```python
df["Salary"] = df["Salary"].str.replace(",", "")
```

###### Convert to numeric:

```python
df["Salary"] = pd.to_numeric(df["Salary"])
```

###### Check

```python
df
```

### Fix Join_Date Column

```python
df["Join_Date"] = pd.to_datetime(df["Join_Date"], errors="coerce")
# Valid dates converted
# invalid_date becomes NaT
```

```python
df
```

### Final Check

```python
df.info()
```

### Save Clean Dataset

```python
df.to_csv("fixed_datatypes_dataset.csv", index=False)
```

## Conclusion

In this task, I learned how to identify and fix incorrect data types in datasets using Pandas. I converted object-type columns into numeric and datetime formats using to_numeric() and to_datetime(). Fixing data types improves data quality and ensures better preprocessing for analysis and machine learning.
